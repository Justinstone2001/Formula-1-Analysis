```python
import pandas as pd
import os

def load_and_clean_data(root_dir):
    df_list = []

    for subdir, _, files in os.walk(root_dir):
        for file in files:
            # Ensure we're looking at CSV files within 'Race Results' directories
            if file.endswith('.csv') and 'Race Results' in subdir:
                file_path = os.path.join(subdir, file)
                try:
                    temp_df = pd.read_csv(file_path)
                    # Extract year from the path
                    year = os.path.basename(os.path.dirname(subdir))
                    temp_df['Year'] = year
                    # Include the filename for reference
                    temp_df['source_file'] = file
                    df_list.append(temp_df)
                except pd.errors.EmptyDataError:
                    print(f"Skipping empty file: {file_path}")

    # Concatenate all valid DataFrames
    if df_list:
        race_results_df = pd.concat(df_list, ignore_index=True)
        
        # Cleaning steps
        columns_to_drop = ['Grand Prix', 'Date', 'Winner', 'Winner Abbreviation', 'Time', 'Q1', 'Q2', 'Q3']
        race_results_df = race_results_df.drop(columns_to_drop, axis=1, errors='ignore')
        
        # Handle missing columns more gracefully
        columns_to_fillna = ['Q1', 'Q2', 'Q3']
        for col in columns_to_fillna:
            if col not in race_results_df.columns:
                race_results_df[col] = 'Not Available'
        
        # Convert numeric columns and fill NaNs
        race_results_df['Driver Number'] = pd.to_numeric(race_results_df['Driver Number'], errors='coerce')
        race_results_df['Laps'] = pd.to_numeric(race_results_df['Laps'], errors='coerce').fillna(0)
        race_results_df['Points'] = pd.to_numeric(race_results_df['Points'], errors='coerce').fillna(0)
        
        return race_results_df
    else:
        print("No valid data files found.")
        return pd.DataFrame()

# Path to your dataset directory
root_dir = 'formula_1_historical_data'
cleaned_race_results_df = load_and_clean_data(root_dir)

# Save to CSV
cleaned_race_results_df.to_csv('cleaned_race_results.csv', index=False)

# Display the first few rows to verify
print(cleaned_race_results_df.head())

cleaned_race_results_df.to_scv('cleaned_results.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Load your cleaned dataset
    cleaned_race_results_df = pd.read_csv('cleaned_race_results.csv')

    # Group the data by 'Car' and 'Year' to aggregate statistics
    car_stats = cleaned_race_results_df.groupby(['Car', 'Year']).agg({'Points': 'sum', 'Driver': 'count'}).reset_index()

    # Calculate the total points earned by each car brand across all years
    car_points = car_stats.groupby('Car')['Points'].sum().sort_values(ascending=False)

    # Determine the number of wins for each car brand
    car_wins = car_stats.groupby('Car')['Driver'].sum().sort_values(ascending=False)

    # Get the top 20 car brands by number of wins
    top_20_car_wins = car_wins.head(20)

    # Plotting
    plt.figure(figsize=(15, 10))

    for car_brand in top_20_car_wins.index:
        # Filter data for the current car brand
        car_data = car_stats[car_stats['Car'] == car_brand]
        
        # Bar plot for total points
        plt.subplot(2, 1, 1)
        plt.bar(car_data['Year'], car_data['Points'], label=car_brand)
        plt.title('Total Points by Car Brand')
        plt.xlabel('Year')
        plt.ylabel('Total Points')
        plt.legend()

        # Bar plot for number of wins
        plt.subplot(2, 1, 2)
        plt.bar(car_data['Year'], car_data['Driver'], label=car_brand)
        plt.title('Number of Wins by Car Brand')
        plt.xlabel('Year')
        plt.ylabel('Number of Wins')
        plt.legend()

    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("The file 'cleaned_race_results.csv' was not found. Please make sure the file exists in the current directory.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set_theme(style="darkgrid")

try:
    # Load your cleaned dataset
    cleaned_race_results_df = pd.read_csv('cleaned_race_results.csv')

    # Preprocess the 'Car' column to extract main brand and engine supplier
    cleaned_race_results_df[['Main Brand', 'Engine Supplier']] = cleaned_race_results_df['Car'].str.split(n=1, expand=True)

    # Define the mappings for brand names
    brand_mapping = {
        # ... (existing mappings)
    }
    
    # Apply the mappings to consolidate the car brands
    cleaned_race_results_df['Main Brand'] = cleaned_race_results_df['Main Brand'].replace(brand_mapping)

    # Group the data by 'Main Brand' and 'Year' to aggregate statistics for wins
    car_wins = cleaned_race_results_df.groupby(['Main Brand', 'Year']).agg({'Driver': 'count'}).reset_index()

    # Filter data for years between 2000 and 2022
    car_wins_filtered = car_wins[(car_wins['Year'] >= 2000) & (car_wins['Year'] <= 2022)]

    # Calculate the total wins earned by each car brand across all years
    top_brands = car_wins_filtered.groupby('Main Brand')['Driver'].sum().nlargest(20).index

    # Filter the data to include only the top brands
    top_brands_data = car_wins_filtered[car_wins_filtered['Main Brand'].isin(top_brands)]

    # Determine the current top three brands based on the latest wins data
    latest_year = top_brands_data['Year'].max()
    latest_wins = top_brands_data[top_brands_data['Year'] == latest_year]
    current_top_three = latest_wins.nlargest(3, 'Driver')['Main Brand']

    # Plotting
    plt.figure(figsize=(15, 10))

    # Plot all brands with lighter lines for total wins
    for brand in top_brands:
        brand_data = top_brands_data[top_brands_data['Main Brand'] == brand]
        if brand not in current_top_three.values:
            plt.plot(brand_data['Year'], brand_data['Driver'], linestyle='--', label=brand, linewidth=1, alpha=0.4)
    
    # Now plot the top three brands with a bolder line for total wins
    for brand in current_top_three:
        brand_data = top_brands_data[top_brands_data['Main Brand'] == brand]
        plt.plot(brand_data['Year'], brand_data['Driver'], marker='o', linestyle='-', label=f"{brand} (Top 3)", linewidth=3)
        # Annotate the latest point
        plt.annotate(f"{brand_data['Driver'].iloc[-1]} wins",
                     (brand_data['Year'].iloc[-1], brand_data['Driver'].iloc[-1]),
                     textcoords="offset points", 
                     xytext=(0,10), 
                     ha='center')

    plt.title('Total Wins by Top 20 Car Brands (2000-2022)')
    plt.xlabel('Year')
    plt.ylabel('Total Wins')
    
    # Improve the legend
    plt.legend(title='Main Brand', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', frameon=True, shadow=True)
    
    # Optional: Simplify the graph by hiding the top and right spines
    sns.despine()

    plt.show()

except FileNotFoundError:
    print("The file 'cleaned_race_results.csv' was not found. Please make sure the file exists in the current directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```
